# <p style="background-color:#4b8cd1;font-family:newtimeroman;color:#151d75;font-size:100%;text-align:center;border-radius:40px 40px;">Outlier Detection</p>

```python
# Use the 'bmh' style for plots
plt.style.use('bmh')

# List of DataFrame columns excluding the last one
df_cols = list(df.columns[:-1])

def CombinePlot(plot_type, columns=df_cols, df=df):
    # Determine the number of rows and columns for subplots
    rows = len(df_cols) // 2
    columns = 2

    # Create a figure and axes for the subplots
    fig, axes = plt.subplots(rows, columns, figsize=(12, 16))

    # Iterate over columns and create the corresponding plots
    for i, col in enumerate(df_cols):
        rows = i // columns
        cols = i % columns

        if plot_type == 'box':
            sns.boxplot(data=df, x=col, ax=axes[rows, cols])
        elif plot_type == 'hist':
            sns.histplot(data=df, x=col, kde=True, ax=axes[rows, cols])

    plt.show()
```

```python
CombinePlot('box')
```

We observe some extreme values, as well as some columns with values in 0. Whose value 0 would be impossible, as in blood pressure or glucose levels. Then we will make a histogram to see a little better the distribution of each column and then impute values to those values 0.

```python
CombinePlot('hist')
```

Now we verify that there are indeed 0 values, in some columns it may be correct, but many others are not (Specifically in 4 of them). Therefore, what we will do is impute the mean of each column to these 0 values.

```python
# Replace 0s with median values in specified columns
zero_values = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for column in zero_values:
    df[column] = df[column].replace(0, df[column].median())
```

Let's visualize the distribution of the columns again once the imputation of the values has been carried out.

```python
CombinePlot('hist', zero_values)
```

`Pregnancies`: We can obtain extreme values. However, when analyzed one by one per person, according to people's ages, it is entirely possible that these extreme values are real.

# <p style="background-color:#4b8cd1;font-family:newtimeroman;color:#151d75;font-size:100%;text-align:center;border-radius:40px 40px;">Data Preprocessing</p>

First of all, then let's remember that our data is unbalanced, so we are going to downsample that majority data (1), which we will do in the next cell.

```python
# Filter rows where 'Outcome' column equals 0 (negative class)
negative_df = df[df['Outcome'] == 0]

# Filter rows where 'Outcome' column equals 1 (positive class)
possitive_df = df[df['Outcome'] == 1]

# Perform undersampling on the negative class
undersampling_df = negative_df.sample(n=268, replace=False, random_state=15)

# Concatenate the undersampled negative class with the positive class
df = pd.concat([undersampling_df, possitive_df])
```

```python
# Extract the 'Glucose' values from the DataFrame
glucose = df.Glucose.values

# Extract the 'Insulin' values from the DataFrame
insulin = df.Insulin.values

# Extract the 'DiabetesPedigreeFunction' values from the DataFrame
pedigree = df.DiabetesPedigreeFunction.values

# Extract the 'BMI' values from the DataFrame
bmi = df.BMI.values
```

```python
# Prepare feature array and target variable
X = np.array([glucose, insulin, bmi, pedigree]).T
y = df.Outcome.values.ravel()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

# <p style="background-color:#4b8cd1;font-family:newtimeroman;color:#151d75;font-size:100%;text-align:center;border-radius:40px 40px;">Metrics</p>

```python
# Evaluate and print classification metrics
def GeneralMetrics(y_true, y_pred):
    print(f'Precision: {metrics.precision_score(y_true, y_pred)}\n')
    print(f'Accuracy: {metrics.accuracy_score(y_true, y_pred)}\n')
    print(f'recall: {metrics.recall_score(y_true, y_pred)}\n')
    print(f'f1_score: {metrics.f1_score(y_true, y_pred)}\n')
    print(f'Area under curve (AUC): {metrics.roc_auc_score(y_true, y_pred)}\n')
    print(f'Confusion matrix: {metrics.confusion_matrix(y_true, y_pred)}\n')
```

# <p style="background-color:#4b8cd1;font-family:newtimeroman;color:#151d75;font-size:150%;text-align:center;border-radius:40px 40px;">Model Selection</p>

```python
# Train and evaluate multiple classification models
def BasicAlgorithms(X_train, y_train):
    models = {
        'LogisticRegression': LogisticRegression(n_jobs=-1),
        'SVC': SVC(random_state=8452),
        'DecisionTreeClassifier': DecisionTreeClassifier(random_state=8452),
        'RandomForestClassifier': RandomForestClassifier(random_state=8452, n_jobs=-1),
        'GradientBoostingClassifier': GradientBoostingClassifier(random_state=8452),
        'AdaBoostClassifier': AdaBoostClassifier(random_state=8452),
        'KNeighborsClassifier': KNeighborsClassifier(n_jobs=-1),
        'GaussianNB': GaussianNB(),
    }
    model_scores = []
    for name, model in models.items():
        # Perform cross-validation
        cv_scores = cross_val_score(model, X_train, y_train, cv=7, scoring='accuracy')
        
        # Calculate evaluation metrics
        accuracy = cv_scores.mean()
        
        # Append the metrics to model_scores
        model_scores.append((name, accuracy))
    
    # Create a DataFrame from model_scores
    columns = ['Model', 'Accuracy']
    df_model_scores = pd.DataFrame(model_scores, columns=columns)
    
    # Sort the DataFrame by Accuracy in descending order
    df_model_scores = df_model_scores.sort_values(by='Accuracy', ascending=False)
    
    # Display the DataFrame
    print(df_model_scores)
```

```python
BasicAlgorithms(X_train, y_train)
```

We are going to build our best model by starting with 3 algorithms, and then determine which is the best of all. First, we will build a random forest, which is the one that had the least accuracy of the 3 we will use. Next, we will use a Gaussian model, which is the one that had the best accuracy. Finally, we will work with logistic regression, whose accuracy is very similar to the Gaussian algorithm.

```python
def RandomForestModel(X_train, y_train, X_test, y_test):
    # Initialize a Random Forest classifier with a fixed random state
    rf = RandomForestClassifier(random_state=15)
    
    # Define hyperparameter grid for tuning
    param_grid = {
        'n_estimators': [200],
        'max_depth': [5],
        'min_samples_split': [7]}
    
    # Perform grid search using cross-validation
    rf = GridSearchCV(estimator=rf, param_grid=param_grid, cv=7, n_jobs=-1)
    rf.fit(X_train, y_train)
    
    # Print the best hyperparameters found during grid search
    print(rf.best_params_)
    
    # Use the best estimator for predictions
    rf = rf.best_estimator_
    y_pred = rf.predict(X_test)
    
    # Evaluate model performance
    GeneralMetrics(y_test, y_pred)
    
    return rf

def LogisticModel(X_train, y_train, X_test, y_test):
    # Define hyperparameter grid for logistic regression
    lr_params = {
        'penalty': ['l2', 'l1', 'elasticnet', 'none'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
        'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
    }

    # Initialize a Logistic Regression classifier
    lr = LogisticRegression(random_state=15)
    
    # Perform grid search using cross-validation
    lr = GridSearchCV(lr, lr_params, cv=7, n_jobs=-1)
    lr.fit(X_train, y_train)

    # Print the best hyperparameters found during grid search
    print(lr.best_params_)
    
    # Use the best estimator for predictions
    y_pred = lr.predict(X_test)
    
    # Evaluate model performance
    GeneralMetrics(y_test, y_pred)
    
    return lr

def GaussianNBModel(X_train, y_train, X_test, y_test):
    # Define hyperparameter grid for Gaussian Naive Bayes
    params_NB = {'var_smoothing': [1e-9, 1e-8, 1e-7], 'priors': [None, [0.2, 0.3, 0.5], [0.1, 0.4, 0.5]]}

    # Initialize a Gaussian Naive Bayes classifier
    nb_classifier = GaussianNB()
    
    # Perform grid search using cross-validation
    nb_classifier = GridSearchCV(estimator=nb_classifier, param_grid=params_NB, cv=7, n_jobs=-1)
    nb_classifier.fit(X_train, y_train)

    # Print the best hyperparameters found during grid search
    print(nb_classifier.best_params_)
    
    # Use the best estimator for predictions
    y_pred = nb_classifier.predict(X_test)
    
    # Evaluate model performance
    GeneralMetrics(y_test, y_pred)
    
    return nb_classifier
```

```python
best_nb = GaussianNBModel(X_train, y_train, X_test, y_test)
```

```python
best_rf = RandomForestModel(X_train, y_train, X_test, y_test)
```

```python
best_lr = LogisticModel(X_train, y_train, X_test, y_test) 
```

```python
# Calculate cross-validated scores using the best logistic regression model (best_lr)
scores = cross_val_score(best_lr, X, y, cv=9)

# Print each individual score
print(scores)

# Calculate and print the average score
print(f'Average score: {scores.mean()}')

# Calculate and print the confidence interval
print(f'Interval confidence: {scores.std() * 2}')

# Print the standard deviation of the scores
print(f'Standard deviation: {scores.std()}')
```

### Model Analysis:
Our best model, then, would definitely be logistic regression. We obtain a better balance between all the metrics, seeing that the model does not "prefer" to throw out any prediction, whether positive or negative, which means that it generalizes very well both what it has learned and the data it has not seen, classifying it in the most stable way possible.

Additionally, as we can observe in the cross-validation, we obtain very similar scores for the metrics, with a very low standard deviation. This indicates that our model does not exhibit overfitting.
